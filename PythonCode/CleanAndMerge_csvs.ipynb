```python
import pandas as pd
from functools import reduce

pd.set_option('display.max_columns', None)
```

```python
#read csvs

lab = pd.read_csv('../Data/CSVs/SA2_LabourStatus.csv')
dist = pd.read_csv('../Data/CSVs/SA2-DistToWork.csv')
comm = pd.read_csv('../Data/CSVs/SA2-CommDevProgPart.csv')
birthc = pd.read_csv('../Data/CSVs/SA2-Birthcountry.csv')
childc = pd.read_csv('../Data/CSVs/SA2-childcare.csv')
corea = pd.read_csv('../Data/CSVs/SA2-CoreActivityAss.csv')
educ = pd.read_csv('../Data/CSVs/SA2-Education.csv')
train = pd.read_csv('../Data/CSVs/SA2-EmploymentTraining.csv')
indus = pd.read_csv('../Data/CSVs/SA2-Industry.csv')
occup = pd.read_csv('../Data/CSVs/SA2-OccupationGroup.csv')
income = pd.read_csv('../Data/CSVs/Sa2-PersonalIncome.csv')
english = pd.read_csv('../Data/CSVs/SA2-ProfEnglish.csv')
marital= pd.read_csv('../Data/CSVs/SA2-RegMaritalStatus.csv')
sector = pd.read_csv('../Data/CSVs/SA2-Sector.csv')
unpaid = pd.read_csv('../Data/CSVs/SA2-UnpaidDomesticWork.csv')
volun = pd.read_csv('../Data/CSVs/SA2-VoluntaryWork.csv')

Allcsvs = ['lab','dist','comm','birthc','childc','corea','educ',
           'train','indus','occup','income','english','marital','sector','unpaid','volun']
```

```python
#function to calculate the total 

def calculatePercentage(df):
    #iterate through columns and calculate percentage for all columns
    for col in df.columns:
        if(col != 'SA2') and (col != 'Total'):
            df[col] = round((df[col] * 100 )/df['Total'],2)
    return df
```

```python
#calculate the percentage for all the dataframes

dist = calculatePercentage(dist)
comm= calculatePercentage(comm)
lab = calculatePercentage(lab)
birthc= calculatePercentage(birthc)
childc= calculatePercentage(childc)
corea= calculatePercentage(corea)
educ= calculatePercentage(educ)
train= calculatePercentage(train)
indus= calculatePercentage(indus)
occup= calculatePercentage(occup)
income= calculatePercentage(income)
english= calculatePercentage(english)
marital= calculatePercentage(marital)
sector= calculatePercentage(sector)
unpaid= calculatePercentage(unpaid)
volun= calculatePercentage(volun)
```
